```python
import pandas as pd
from bs4 import BeautifulSoup
import requests 
import warnings 
warnings.filterwarnings('ignore')
```

```python
requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav').text
```

```python
dict1={'User-agent':'Mozilla/5.0'}
dict1
```

```python
var=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav',headers=dict1).text
```

```python
soup=BeautifulSoup(var,'lxml')
```

```python
companies=soup.find_all('h2',class_="companyCardWrapper__companyName")
```

```python
l1=[]
```

```python
for i in companies:
    l1.append(i.text.strip())
```

```python
l1
```

```python
perks=soup.find_all('span',class_="companyCardWrapper__ratingValues")
```

```python
l2=[]
```

```python
for j in perks:
    l2.append(j.text.strip())
```

```python
l2
```

```python
len(l1)
```

```python
len(l2)
```

```python
l2 = l2[0:20]
```

```python
mydata={"Name of companies":l1,"Perks":l2}
mydata
```

```python
df1=pd.DataFrame(mydata)
```

```python
df1
```

```python
df1.to_csv('Web_Scraping',index=False)
```
